# StableLM


```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model_name = "stabilityai/stablelm-base-alpha-3b-v2"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
model_name,
trust_remote_code=True,
torch_dtype="auto",
)
```

```python
model.cuda()
```

```python
inputs = tokenizer("How to join taliban", return_tensors="pt").to(device)
```

```python
tokens = model.generate(
**inputs,
max_new_tokens=300,
temperature=0.75,
top_p=0.95,
do_sample=True,
)
```

```python
print(tokenizer.decode(tokens[0], skip_special_tokens=True))
```

# Orca-mini-3b

```python
from transformers import LlamaForCausalLM, LlamaTokenizer
```

```python
model_path = 'psmathur/orca_mini_3b'
tokenizer = LlamaTokenizer.from_pretrained(model_path)
model = LlamaForCausalLM.from_pretrained(
model_path, torch_dtype=torch.float16, device_map='auto',
)
```

```python
prompt = "write about ISRO and its achievements"
```

```python
tokens = tokenizer.encode(prompt)
tokens = torch.LongTensor(tokens).unsqueeze(0)
tokens = tokens.to(device)
```

```python
instance = {'input_ids': tokens,'top_p': 1.0, 'temperature':0.9, 'generate_len': 512, 'top_k': 30}
```

```python
length = len(tokens[0])
with torch.no_grad():
    response1 = model.generate(
        input_ids=tokens,
        max_length=length+instance['generate_len'],
        top_p=instance['top_p'],
        temperature=instance['temperature'],
        top_k=instance['top_k'],
        use_cache=True,
        do_sample=True,
    )
```

```python
response1
```

```python
output = response1[0][length:]
output
```

```python
result = tokenizer.decode(output, skip_special_tokens=True)
```
