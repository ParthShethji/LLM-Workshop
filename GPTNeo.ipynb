```python
from transformers import BertTokenizerFast, GPT2LMHeadModel
```

```python
tokenizer = BertTokenizerFast.from_pretrained("kykim/gpt3-kor-small_based_on_gpt2")
input_ids = tokenizer.encode("text to tokenize")[1:]

# takes a piece of text (in this case, "text to tokenize") and converts it into numerical data, represented by a sequence of IDs.
# Each unique word or subword in the text is assigned a unique ID,
```
